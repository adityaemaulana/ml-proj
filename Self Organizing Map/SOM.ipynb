```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
data = np.genfromtxt('Dataset.csv', delimiter=',')
```

```python
plt.scatter(x=data[:,0], y=data[:,1])
```

```python
# Initialization

# Create random nxn matrices for output layer
# Each cell have 2 random value [0,1]

output_layer = np.random.random_sample(size=(17,17,2))

# Set learning rate
learning_rate = 0.1

# Set width variance
width = 2

# Set epoch
epoch = 1000
```

```python
# Calculate euclidean distance
def euclidean(x, y):
    return math.sqrt((x[0]-y[0])**2 + (x[1]-y[1])**2)

# Neighborhood Function to calculate how close is two node
def neighborhood(dist, width):
    return math.exp(-dist / (2 * (width ** 2)))

# Update neighborhood
def update_neighborhood(width, iteration, epoch):
    neighborhood = width * math.exp(-iteration / epoch)
    if(neighborhood <= 0):
        return width
    else:
        return neighborhood

# Update learning rate
def update_learning_rate(learning_rate, iteration, epoch):
    return learning_rate * math.exp(-iteration / epoch)

def find_bmu(x):
    # Find Best Matching Unit(BMU) of each input
    bmu = []
    min_dist = 999999

    for i in range(len(output_layer)):
        for j in range(len(output_layer)):
            curr_dist = euclidean(x, output_layer[i][j])
            if(curr_dist < min_dist):
                min_dist = curr_dist
                bmu = output_layer[i][j]

    return bmu
```

```python
plt.scatter(x=output_layer[:,:,0], y=output_layer[:,:,1], color='blue')
```

```python
indices = np.arange(len(data))

for i in range(1, epoch + 1):
    print("#{}..............".format(i))
    print("width: {}".format(width))
    np.random.shuffle(indices)

    # Iterasi seluruh isi dari dataset
    for idx in indices:
        x = data[idx]
        
        # Cari node pada output layer yang paling 'dekat' dengan x

        bmu = find_bmu(x)
        
        # Update bobot seluruh node yang 'dekat' dengan BMU
        for row in range(len(output_layer)):
            for col in range(len(output_layer)):
                curr_node = output_layer[row][col]
                curr_dist = euclidean(bmu, curr_node)

                # Apabila node ini berdekatan dengan BMU
                if(curr_dist <= width**2):
                    delta_weight = learning_rate * neighborhood(curr_dist, width) * (x - curr_node)
                    output_layer[row][col] += delta_weight
    plt.figure()
    plt.scatter(x=output_layer[:,:,0], y=output_layer[:,:,1], color='blue')
    plt.show()
    # Update Hyperparameter          
    width = update_neighborhood(width, i, epoch)
    learning_rate = update_learning_rate(learning_rate, i, epoch)
```
